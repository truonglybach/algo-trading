# Observational Study of Sentimental Analysis

The aim of my project is to find correlations between the sentiment of various stocks and their corresponding daily returns grouped by similarity (market cap, stock price, etc.) and varying by sector. I will also develop a forecasting model using a LSTM deep learning network, which will base its predictions on time series data in order to gain daily returns.

```python
# Importing basic dependencies
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from datetime import datetime
from scipy.stats import linregress

# Research environment functions
from quantopian.research import prices, returns, symbols

# Pipeline imports
from quantopian.research import run_pipeline
from quantopian.pipeline import Pipeline
from quantopian.pipeline.factors import Returns
from quantopian.pipeline.data.psychsignal import stocktwits
```

# Exploring the Data

```python
# A function that grabs today's date in a specific format (Year-month-day)
def grab_today():
    return datetime.now().strftime('%Y-%m-%d')
today = grab_today()
```

A note regarding which dates to use:

Although one should update his/her data with recent prices, we will used a fixed set of dates to conduct our study for consistency in our outcome since our model is heavily reliant on historical data in order to make statistical inferences.

```python
# Create arbitary start and end dates for data exploration
start_period = '2016-01-15'
end_period = '2019-01-15'
```

```python
# A function that queries the closing prices for a given stock
def closing_prices(str_ticker, start_date=start_period, end_date=end_period):
    return prices(
    assets=symbols(str_ticker),
    start=start_date,
    end=end_date, 
    price_field='close'
)

# Query returns data for a given stock over the selected time range
def returns_data(str_ticker, start_date=start_period, end_date=end_period):
    return returns(
    assets=symbols(str_ticker),
    start=start_date,
    end=end_date,
)
```

```python
# Grab the closing prices (from January 15, 2016 to January 15, 2019) for Advaxis
adxs_close = closing_prices('ADXS')

# Compute 20 and 50 day simple moving averages on pricing data
adxs_sma20 = adxs_close.rolling(20).mean()
adxs_sma50 = adxs_close.rolling(50).mean()

# Combine results into a pandas DataFrame and plot
pd.DataFrame({   
    'ADXS': adxs_close,
    'SMA20': adxs_sma20,
    'SMA50': adxs_sma50
}).plot(
    title='ADXS Close Price / SMA Crossover'
);
```

```python
# Pipeline definition
def make_pipeline():
    # DailyReturns calculates the daily percent change in close price
    # (min window_length is 2)
    returns = Returns(window_length=2)
    sentiment = stocktwits.bull_minus_bear.latest
    msg_volume = stocktwits.total_scanned_messages.latest

    return Pipeline(
        columns={
            'daily_returns': returns,
            'sentiment': sentiment,
            'msg_volume': msg_volume,
        },
    )

# Pipeline execution
data_output = run_pipeline(
    make_pipeline(),
    start_date=start_period,
    end_date=end_period
)
```

```python
# Filter results for ADXS
adxs_output = data_output.xs(
    symbols('ADXS'),
    level=1
)

# Plot results for ADXS
adxs_output.plot(subplots=True);
```

### Note on a calculation discrepancy:
This discrepancy is in regards to the exact details in which dates are used to calculate the
daily return as a percentage for a given day, x:
The dates that are used to calculate the daily return for x can be viewed as
(x-1) - (x-2) / x-2.

For example, if we were to calculate the daily return for January 3rd, then we would take the closing price on January 2nd and subtract that from the preceding closing price on December 31st (because NASDAQ does not operate on some holidays) and calculate for the percentage gain/loss for that time period.

https://www.kiplinger.com/article/investing/T038-C000-S001-2019-stock-market-holidays-bond-market-holidays.html

```python
# To note the discrepancy:
adxs_close.tail(11)
```

```python
DR = (adxs_close.loc['2019-01-02 00:00:00+00:00'] - adxs_close.loc['2018-12-31 00:00:00+00:00']) / adxs_close.loc['2018-12-31 00:00:00+00:00']
DR
```

We calculated the percentage gain of January 2nd from December 31st, but it's showing up on January 3rd.

For our purposes, I will define 'daily return' as if I had invested at the closing price the day before and sold at the closing price on the day of; therefore, I will need to shift my data up a period.

```python
adxs_output['daily_returns'] = adxs_output['daily_returns'].shift(periods=-1)
```

```python
adxs_output['daily_returns'].tail(10)
```

Now let's try to draw a line of best fit between our daily returns and sentiment scores:

```python
# Grab x and y
x = adxs_output['sentiment']
y = adxs_output['daily_returns']

# Plot sentiment against daily returns
adxs_output[['sentiment', 'daily_returns']].plot(x='sentiment', y='daily_returns', kind='scatter', title='Sentiment versus Daily Returns')

# Calculate linear regression model based on x and y values
slope, intercept, r_value, _, _ = linregress(x, y)

# Plot line of best fit
x_bf = np.linspace(-4, 4)
y_bf = slope * x_bf + intercept
line_best_fit = plt.plot(x_bf, y_bf, 'r--')
```

```python
print('The correlation coefficient (r = {}) for these two variables suggest that the relationship between any movement in x is very weak in regards to any movement in y.').format(r_value)
r2 = r_value ** 2
print('')
print('Our coefficient of determination (r^2 = {}) tells us that the percentage of the total variation of y is very unlikely to be described by the total variation of x, in other words, it would be very unlikely for the spread of our independent variable to explain the spread of our dependent variable.').format(r2)
```

Although our data shows us that there is a very weak correlation between our two variables, perhaps we can find a find a better r-value and r^2 if we narrow the search.

To paraphrase, they state that stock prices are more responsive to negative sentiment within the scope of intraday trading than positive sentiment.

If this was the case, we should see a stronger correlation between daily losses and negative sentiment than just comparing daily returns to general sentiment.

We'll grab all of the necessary data and store it into one dataframe.

```python
# Takes in a list strings and iterates through each stock ticker
def grab_stock_data(stock_list):
    # an empty dataframe
    concat_df = pd.DataFrame(None)
    # iterating over every stock symbol
    for i in stock_list:
        # creating empty lists for multilevel index
        axis_0 = []
        axis_1 = []
        # appending ticker symbol twice
        axis_0.append(i)
        axis_0.append(i)
        # takes a cross-section original multilevel dataframe above
        # and stores the output
        output = data_output.xs(
            symbols(i),
            level=1
            )
        # appends the info we want to our empty list
        axis_1.append('daily_returns')
        axis_1.append('sentiment')
        # stores sentiment values
        x = output['sentiment']
        # stores daily_returns
        y = output['daily_returns'].shift(periods=-1)
        # grabs the extent of the shape
        x_shaper = x.shape[0]
        y_shaper = y.shape[0]
        # changes the shape of x and y in prep of multilevel dataframe
        daily_returns = np.array(x).reshape(1, x_shaper)
        sentiment = np.array(y).reshape(1, y_shaper)
        # creates a list of the zipped index values
        tuples = list(zip(axis_0, axis_1))
        # concatenates our two arrays
        data = np.concatenate((daily_returns, sentiment), axis=0)
        # creates the index from our list of tuples
        index = pd.MultiIndex.from_tuples(tuples, names=['Ticker', 'Data'])
        # turns our grabbed data into a dataframe
        result = pd.DataFrame(data, index)
        # concatenates the dataframe to our empty df above
        concat_df = pd.concat([concat_df, result], axis=0)
    # returns the finished df
    return concat_df
```

# Nevermind!
We -ARE NOT- able to impute our data efficiently. If we did it by hand, we would have to use our eyes to catch every 'nan' in our data frame and manually replace that with the median of each list (which is ridiculous).

What we can still do is simply iterate through our list of stocks again and create a data frame that holds the necessary information in order to see if there are any significant correlations existing within our list.

We will iterate through our list for a total of two times. The first instance we will grab all values (not caring about positive or negative sentiment), and the second, we will limit our search to values where sentiment is negative.

```python
# Define a function that iterates through our list of stocks and creates a
# pandas data frame with the grabbed data
def grab_r_values_all_sentiment(stock_list):
    data = []
    for i in stock_list:
        output = data_output.xs(
            symbols(i),
            level=1
        )
        x = output['sentiment']
        y = output['daily_returns'].shift(periods=-1)
        x = x[:len(y)-1]
        y = y[:len(y)-1]
        slope, intercept, r_value, _, _ = linregress(x, y)
        data.append(r_value)
    return pd.DataFrame(data=data, columns=['r-value'])

# Define a function that iterates through our list of stocks and creates a
# pandas data frame with the grabbed data that is limited to negative sentiment
def grab_r_values_neg_sentiment(stock_list):
    data = []
    for i in stock_list:
        output = data_output.xs(
            symbols(i),
            level=1
        )
        # Filter out neutral & positive scores
        x = output['sentiment']
        y = output['daily_returns'].shift(periods=-1)
        x = x.loc[output['sentiment'] < 0][:len(y)-1]
        y = y.loc[output['sentiment'] < 0][:len(y)-1]
        slope, intercept, r_value, _, _ = linregress(x, y)
        data.append(r_value)
    return pd.DataFrame(data=data, columns=['r-value'])
```

```python
# Store the r-values in a variable
r_values_all = grab_r_values_all_sentiment(sim_stocks)
r_values_neg = grab_r_values_neg_sentiment(sim_stocks)
```

```python
def r_score_r2_df_maker(r_values, stock_list):
    # Put the stock names in a data frame
    sim_stocks_df = pd.DataFrame(stock_list, columns=['sim_stocks'])
    # Add the values to the data frame
    sim_stocks_df['r_values'] = r_values
    # Drop all null values (So stocks with missing r-values will be dismissed)
    sim_stocks_df.dropna(inplace=True)
    # Add a column that takes the r^2 values and store
    sim_stocks_df['r^2'] = sim_stocks_df['r_values'] ** 2
    return sim_stocks_df
```

# Forecasting Model

After conducting research over the issue, I have decided that although the r^2 scores were significant, basing an alpha model that's supposed to trade daily on forecasted sentiment values wouldn't be a good idea because we don't have consistent input from a day to day basis.

Instead, I will develop a LSTM network to predict fluctuations in stock prices since they work great with sequential data where you'd want it to remember the important trends and forget the nonsense.

In the article linked down below, the author compares several algorithms against each other and determined that an LSTM network was the best algorithm suited for this purpose:

https://www.analyticsvidhya.com/blog/2018/10/predicting-stock-price-machine-learningnd-deep-learning-techniques-python/

Unfortunately, as stated above, we will not be able to run our necessary dependencies and download the data through the Quantopian Research platform; therefore, I will use Quandl to get the necessary data and run the LSTM network locally.
